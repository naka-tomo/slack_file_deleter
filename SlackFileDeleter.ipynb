```python
import requests
import json
from datetime import datetime
```

# 設定

```python
# slackのwebページで生成したtoken
token = ""

# この日付より前のファイルを削除する
time =  datetime.strptime( "2019-03-31", "%Y-%m-%d" )

# 時刻をunix時間に変換
time = time.timestamp()
```

# 関数定義

```python
def get_file_list( ts_to=None ,channel=None ):
    file_list = []
    url = "https://slack.com/api/files.list?token=" + token

    if ts_to:
        url +=  "&ts_to=" + str(ts_to)
    
    if channel:
        url += "&channel=" + channel
    
    r = requests.get(  url )

    data = json.loads(r.text)
    
    N = len(data["files"])
    for  n in range(N):
        name = data["files"][n]["name"] 
        id = data["files"][n]["id"] 
        file_list.append( (name, id) )
    
    return file_list

def delete_file( id ):
    r = requests.get(  "https://slack.com/api/files.delete?token=" + token + "&file=" + id  )
    data = json.loads(r.text)
    return data["ok"]

def get_channel_list():
    dchannle_dic = {}
    r = requests.get(  "https://slack.com/api/channels.list?token=" + token )
    data = json.loads(r.text)
    
    N = len(data["channels"])
    for n in range(N):
        name = data["channels"][n]["name"]
        id = data["channels"][n]["id"]
        dchannle_dic[name] = id
    return dchannle_dic
```
